```python
#importing stuff
import numpy as np
import pandas as pd
import json
import os #used to cycle through your files
import re
```

```python
# reading in data from 
#   https://www.kaggle.com/datasets/mariotormo/complete-pokemon-dataset-updated-090420

pokedex = pd.read_csv('pokedex.csv')
```

```python
pokedex
```

## Getting rid of extra columns

```python
#dropping extra index column
pokedex.drop(pokedex.columns[0], axis=1,inplace=True)
```

```python
#dropping German name
pokedex.drop(labels='german_name', axis=1, inplace=True)
```

```python
#dropping Japanese name
pokedex.drop(labels='japanese_name', axis=1, inplace=True)
```

```python
#dropping species
pokedex.drop(labels='species', axis=1, inplace=True)
```

```python
#looking to see if I really need/want the type_number column
pokedex['type_2'].unique()

#decided to drop type_number column
pokedex.drop(labels='type_number', axis=1, inplace=True)
```

```python
#finding egg type columns
pokedex.iloc[:, 23:26]

#dropping egg_type_number, egg_type_1, and egg_type_2
pokedex.drop(pokedex.columns[23:26], axis=1, inplace=True)
```

```python
#finding ability type columns
pokedex.iloc[:, 8:12]

#dropping ability_number, ability_one, ability_two, and ability_hidden
pokedex.drop(pokedex.columns[8:12], axis=1, inplace=True)
```

```python
#getting rid of special attack and defense columns
pokedex.drop(labels='sp_attack', axis=1, inplace=True)
pokedex.drop(labels='sp_defense', axis=1, inplace=True)
```

```python
#dropping percentage_male
pokedex.drop(labels='percentage_male', axis=1, inplace=True)
```
